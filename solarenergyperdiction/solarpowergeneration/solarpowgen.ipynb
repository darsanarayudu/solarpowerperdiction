```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
df = pd.read_csv("solarpowergeneration.csv")
# Display the first few rows to ensure the data loaded correctly
print(df.head())
# Check for missing values in the 'generated_power_kw' column
missing_values = df['generated_power_kw'].isnull().sum()
print(f"Missing values in 'generated_power_kw': {missing_values}")
# Handle missing values 
#Fill missing values with the column mean
df['generated_power_kw'] = df['generated_power_kw'].fillna(df['generated_power_kw'].mean())
# Option 2: Alternatively, you can drop rows with missing values if it's more appropriate
df = df.dropna(subset=['generated_power_kw'])
# Plot the distribution of solar power generation
plt.figure(figsize=(10, 8))
sns.histplot(df['generated_power_kw'], bins=30, kde=True, color='skyblue', line_kws={'color': 'red', 'linewidth': 2})
# Enhance plot with titles and labels
plt.title("Distribution of Solar Power Generation (kW)", fontsize=16, weight='bold')
plt.xlabel("Generated Power (kW)", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
# Show grid for better readability
plt.grid(True)
# Optional: Limit the x-axis to remove extreme outliers (if applicable)
plt.xlim(df['generated_power_kw'].min(), df['generated_power_kw'].quantile(0.95))
plt.show()
# Display summary statistics of the 'generated_power_kw' column to understand its distribution
print(df['generated_power_kw'].describe())
#Optional: Correlation heatmap to identify relationships between features (if dataset has multiple columns)
corr_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```
